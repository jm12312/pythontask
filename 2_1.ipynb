```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("netflix_titles.csv")
data = data.dropna()
data_filtered = data.reset_index()
data_filtered
```

```python
data_filtered.head(10)
```

```python
exp = (data_filtered["release_year"] == 2021) & (data_filtered["type"] == "Movie")
df1 = data_filtered[exp].reset_index()
df1[["release_year", "type"]]
```

```python
exp2 = ((data_filtered['type']=='Movie') & (data_filtered['title'].str.contains('Avengers')))
df2 = data_filtered[exp2].reset_index()
df2[["title"]]
```

```python
filt2 = (data_filtered["type"] == "Movie")
df4 = data_filtered.loc[filt2]
df4.reset_index(inplace=True)
list_all = df4["country"].str.split(', ')


list_all
```

```python

list_all = list_all.to_numpy()
list_all
```

```python
list_unique = np.unique(sum(list_all, []), return_counts=False)[1:]

list_unique
```

```python
list_value = [0] * len(list_unique)
for i in range(len(list_unique)):
    for j in range(len(df4)):
        if list_unique[i] in df4.iloc[j]["country"]:
            list_value[i] += 1
print(list_value)
```

```python
dict1 = {list_unique[i]: list_value[i] for i in range(len(list_value))}
print(dict1)
```

```python
x =pd.DataFrame(dict1.items())
x.drop(104, inplace=True)
x.drop(78, inplace=True)
x.drop(15, inplace=True)
x = x.reset_index()

x.tail(40)
```

```python
x=x.sort_values(by=[1], ascending=False)
x.head(10)
```

```python
x=x[[0,1]][:5]
x.reset_index(inplace=True)
x
```

```python
separate = [0.2]*len(x)
plt.pie(x[1], labels=x[0], autopct='%1.1f%%', explode=separate)
plt.title("Movies per country")

plt.show()
```

```python
#TV Shows
filt3 = (data_filtered["type"] == "TV Show")
df5 = data_filtered.loc[filt3]
df5.reset_index(inplace=True)
list_all1 = df5["country"].str.split(', ').to_numpy()
list_unique1 = np.unique(sum(list_all1, []), return_counts=False)[1:]

list_unique1
```

```python
list_value1 = [0] * len(list_unique1)
for i in range(len(list_unique1)):
    for j in range(len(df5)):
        if list_unique1[i] in df5.iloc[j]["country"]:
            list_value1[i] += 1
print(list_value1)
```

```python
dict2 = {list_unique1[i]: list_value1[i] for i in range(len(list_value1))}
x1 =pd.DataFrame(dict2.items())
x1=x1.sort_values(by=[1], ascending=False)
x1.reset_index(inplace=True)
x1.head(10)
```

```python
x1=x1[[0,1]][:5]
x1
```

```python
separate1 = [0.2]*len(x1)
plt.pie(x1[1], labels=x1[0], autopct='%1.1f%%', explode=separate1)
plt.title("TV Shows per country")

plt.show()
```
